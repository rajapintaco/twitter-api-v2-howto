## Demo: Collecting tweets for a user

Fetching tweets for a particular user by using twarc's <code>timeline</code> command.

```python
!twarc timeline rajapintaco > rajapinta-user-tweets.jsonl
```

```python
!ls 
```

```python
import json

with open('rajapinta-user-tweets.jsonl') as f:
    tweets_jsonl = f.read()
    
tweets = [json.loads(jline) for jline in tweets_jsonl.splitlines()]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(tweets)
```
